```python
# Scatter and density plots
def plotScatterMatrix(df, plotSize, textSize):
    df = df.select_dtypes(include =[np.number]) # keep only numerical columns
    # Remove rows and columns that would lead to df being singular
    df = df.dropna('columns')
    df = df[[col for col in df if df[col].nunique() > 1]] # keep columns where there are more than 1 unique values
    columnNames = list(df)
    if len(columnNames) > 10: # reduce the number of columns for matrix inversion of kernel density plots
        columnNames = columnNames[:10]
    df = df[columnNames]
    ax = pd.plotting.scatter_matrix(df, alpha=0.75, figsize=[plotSize, plotSize], diagonal='kde')
    corrs = df.corr().values
    for i, j in zip(*plt.np.triu_indices_from(ax, k = 1)):
        ax[i, j].annotate('Corr. coef = %.3f' % corrs[i, j], (0.8, 0.2), xycoords='axes fraction', ha='center', va='center', size=textSize)
    plt.suptitle('Scatter and Density Plot')
    plt.show()
```

####### Looking at the model building part performed above, Random Forest Classifier will be a best model to predict the accuracy score that a new wine will recieve. 

###### Advantages of using Random Forest Classifier are: 
#######    1. It reduces overfitting in decision trees, also reduces the variance and helps to improve the accuracy.
#######    2. It is flexible to both classification and regression problems.
#######    3. It works well with both categorical and continuous values.
#######    4. It automates missing values and outliers present in the data.
#######    5. Normalising of data is not required as it uses a rule-based approach.
#######    6. Random Forest is comparatively less impacted by noise.

###### Disadvantages of using Random Forest Classifier are:
#######    1. It requires much time for training as it combines a lot of decision trees to determine the class.
#######    2. Random forest requires much computational power as well as resources as it builds numerous trees to combine their outputs. 
#######    3. Due to the ensemble of decision trees, it also suffers interpretability and fails to determine the significance of each variable.
